```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load data
df = pd.read_csv("../data/fake_job_postings.csv")  # Adjust path if needed
df_clean = df[['description', 'fraudulent']].dropna()

# --- Fraudulent Word Cloud ---
fraud_text = " ".join(df_clean[df_clean['fraudulent'] == 1]['description'])
fraud_wc = WordCloud(width=800, height=400, background_color='white', random_state=42).generate(fraud_text)

plt.figure(figsize=(10, 5))
plt.imshow(fraud_wc, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud: Fraudulent Job Postings")
plt.tight_layout()
plt.savefig("../wordclouds/wordcloud_fraudulent.png")  # Adjust if saving elsewhere
plt.show()

# --- Real Word Cloud ---
real_text = " ".join(df_clean[df_clean['fraudulent'] == 0]['description'])
real_wc = WordCloud(width=800, height=400, background_color='white', random_state=42).generate(real_text)

plt.figure(figsize=(10, 5))
plt.imshow(real_wc, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud: Real Job Postings")
plt.tight_layout()
plt.savefig("../wordclouds/wordcloud_real.png")  # Adjust if saving elsewhere
plt.show()
```
